# Train test split

```python
# first we need to dumify the data
df = pd.get_dummies(data = df , columns= cats , drop_first = True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(['Converted'] , axis=1 ), df['Converted'], train_size=0.7, test_size=0.3, random_state=24)
```

# Modeling

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.feature_selection import RFE
```

```python
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
```

```python
preds = clf.predict(X_test)
```

```python
cm = confusion_matrix(y_test, clf.predict(X_test))

fig, ax = plt.subplots(figsize=(6, 6))
ax.imshow(cm)
ax.grid(False)
ax.xaxis.set(ticks=(0, 1), ticklabels=('Predicted 0s', 'Predicted 1s'))
ax.yaxis.set(ticks=(0, 1), ticklabels=('Actual 0s', 'Actual 1s'))
ax.set_ylim(1.5, -0.5)
for i in range(2):
    for j in range(2):
        ax.text(j, i, cm[i, j], ha='center', va='center', color='red')
plt.show()
```

```python
print(classification_report(y_test, clf.predict(X_test))) # looks good lets see if RFE could help us
```

```python
# lets plot some graphs
```

```python
from sklearn.metrics import roc_curve       
from sklearn.metrics import roc_auc_score
```

```python
def plot_roc( real, probability ):
    #getting the necessary values
    fpr, tpr, thresholds = roc_curve( real, probability,drop_intermediate = False )   
    auc_score = roc_auc_score( real, probability )   #auc  
    plt.figure(figsize=(12, 6))    
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )     
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC')
    plt.legend(loc="lower right")
    plt.show()
```

```python
plot_roc(y_test,  clf.predict_proba(X_test)[:,1])
```

```python
p ,r, thresholds=precision_recall_curve(y_test,clf.predict_proba(X_test)[:,1])
plt.title('Precision vs Recall tradeoff')
plt.plot(thresholds, p[:-1], "g-")    # Plotting precision
plt.plot(thresholds, r[:-1], "r-")    # Plotting Recall
plt.show()
```

the tradeoff is around 0.4 lets use that as threshold
